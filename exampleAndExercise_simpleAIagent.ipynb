```python
# Simple AI Agent for Booking Appointments

import datetime

# Mock schedule (some booked times)
mock_schedule = {
    "doctor": ["2025-05-28 10:00", "2025-05-28 14:00"],
    "dentist": ["2025-05-29 09:00"],
    "haircut": []
}

def is_available(appointment_type, datetime_str):
    """Check if the requested time is available."""
    return datetime_str not in mock_schedule.get(appointment_type, [])

def book_appointment(appointment_type, date_str, time_str):
    datetime_str = f"{date_str} {time_str}"
    if is_available(appointment_type, datetime_str):
        mock_schedule[appointment_type].append(datetime_str)
        return f"‚úÖ Your {appointment_type} appointment is booked for {datetime_str}."
    else:
        return f"‚ùå Sorry, the {appointment_type} appointment at {datetime_str} is already booked. Please choose another time."

# Simulated conversation
def ai_agent():
    print("ü§ñ Hello! I'm your appointment assistant.")
    appt_type = input("What type of appointment would you like to book (doctor, dentist, haircut)? ").lower()

    if appt_type not in mock_schedule:
        print("Sorry, I can't book that type of appointment.")
        return

    date_str = input("What date would you prefer? (YYYY-MM-DD) ")
    time_str = input("What time would you like? (HH:MM in 24h format) ")

    try:
        # Validate date and time
        datetime.datetime.strptime(f"{date_str} {time_str}", "%Y-%m-%d %H:%M")
        result = book_appointment(appt_type, date_str, time_str)
        print(result)
    except ValueError:
        print("‚ùó Invalid date or time format. Please try again.")

# Run the agent
if __name__ == "__main__":
    ai_agent()
```
