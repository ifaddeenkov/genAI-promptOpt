## Exercise 3: transform the following code from R to Python using genAI

### (Expected Results can checked on the Github repository)



```
# Load necessary libraries
library(ggplot2)
library(dplyr)

# 1. Simulate data
set.seed(42)
df <- data.frame(
  group = rep(c("Control", "Treatment_A", "Treatment_B"), each = 40),
  value = c(rnorm(40, mean = 5, sd = 1),
            rnorm(40, mean = 6, sd = 1.2),
            rnorm(40, mean = 7, sd = 1.5))
)

# 2. Preprocess: Ensure factor type
df$group <- as.factor(df$group)

# 3. Statistical Tests
# Normality check per group
by(df$value, df$group, shapiro.test)

# Bartlett's test for homogeneity of variances
bartlett.test(value ~ group, data = df)

# Kruskal-Wallis test (non-parametric ANOVA alternative)
kruskal <- kruskal.test(value ~ group, data = df)
print(kruskal)

# Pairwise Wilcoxon test with Bonferroni correction
pairwise <- pairwise.wilcox.test(df$value, df$group, p.adjust.method = "bonferroni")
print(pairwise)

# 4. Compute summary statistics for plotting
summary_df <- df %>%
  group_by(group) %>%
  summarise(
    median = median(value),
    q1 = quantile(value, 0.25),
    q3 = quantile(value, 0.75)
  )

# 5. Visualization
p <- ggplot(df, aes(x = group, y = value, fill = group)) +
  geom_boxplot(alpha = 0.6, outlier.shape = NA) +
  geom_jitter(width = 0.2, alpha = 0.4, size = 1.2) +
  stat_summary(fun = median, geom = "point", shape = 21, size = 3, fill = "black") +
  labs(
    title = "Group Comparison: Kruskal-Wallis Test",
    subtitle = paste("p =", signif(kruskal$p.value, 3)),
    x = "Group", y = "Response"
  ) +
  theme_minimal(base_size = 14) +
  theme(legend.position = "none")

# Optional: Add asterisks manually based on post-hoc p-values
# Example: suppose significant p < 0.05 for A vs B and Control vs B
p + annotate("text", x = 1.5, y = 10.5, label = "*", size = 6) +
    annotate("text", x = 2.5, y = 11.0, label = "**", size = 6)
```
